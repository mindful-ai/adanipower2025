```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
boiler_input = pd.read_csv('boilerinput.csv')
boiler_output = pd.read_csv('boileroutput.csv')
O2_dataset = pd.read_csv('O2_dataset.csv')
coal_df = pd.read_csv('coaldf.csv')


```

```python
# Display dataset information
def dataset_overview():
    print("Boiler Input Dataset Overview:")
    print(boiler_input.info(), "\n")
    print("Boiler Output Dataset Overview:")
    print(boiler_output.info(), "\n")
    print("O2 Dataset Overview:")
    print(O2_dataset.info(), "\n")
    print("Coal Dataset Overview:")
    print(coal_df.info(), "\n")

```

```python
# Summary statistics
def summary_statistics():
    print("Summary of Boiler Input Dataset:")
    print(boiler_input.describe(), "\n")
    print("Summary of Boiler Output Dataset:")
    print(boiler_output.describe(), "\n")
    print("Summary of O2 Dataset:")
    print(O2_dataset.describe(), "\n")
    print("Summary of Coal Dataset:")
    print(coal_df.describe(), "\n")

```

```python
# Visualization examples
def create_visualizations():
    plt.figure(figsize=(12, 6))
    sns.histplot(boiler_input['APH-B O/L O2'], kde=True, bins=20, color='blue')
    plt.title('Distribution of APH-B O/L O2 in Boiler Input Dataset')
    plt.xlabel('APH-B O/L O2')
    plt.ylabel('Frequency')
    plt.show()

    plt.figure(figsize=(12, 6))
    sns.scatterplot(x=coal_df['GCV'], y=coal_df['Carbon In Fuel'], hue=coal_df['Moisture in Fuel'], palette='cool')
    plt.title('Relationship between GCV, Carbon In Fuel, and Moisture in Fuel')
    plt.xlabel('GCV')
    plt.ylabel('Carbon In Fuel')
    plt.show()

    plt.figure(figsize=(12, 6))
    sns.lineplot(x='time', y='Boiler Efficiency', data=boiler_output, marker='o')
    plt.title('Boiler Efficiency Over Time')
    plt.xlabel('Time')
    plt.ylabel('Boiler Efficiency')
    plt.xticks(rotation=45)
    plt.show()

    plt.figure(figsize=(12, 6))
    sns.heatmap(O2_dataset.corr(), annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Heatmap of O2 Dataset')
    plt.show()

```
